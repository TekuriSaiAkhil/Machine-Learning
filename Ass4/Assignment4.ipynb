```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import preprocessing
from sklearn import svm
from sklearn.datasets import fetch_kddcup99
from sklearn.preprocessing import OneHotEncoder
from sklearn.model_selection import train_test_split
from imblearn.under_sampling import RandomUnderSampler
from sklearn.inspection import DecisionBoundaryDisplay
```

```python
#Loading dataset and cleaning
df = pd.read_csv('kddcup99_csv.csv')
df.drop(df[df['label'] == 'normal'].index, inplace = True)
df = df.drop_duplicates()
df = df.dropna()
```

```python
#One hot encoding of categorical features
categories_data = df[['protocol_type','service','flag']]
enc = OneHotEncoder(handle_unknown='ignore')
enc.fit(categories_data)
categories_data_transformed = enc.transform(categories_data).toarray()
categories_data_transformed = pd.DataFrame(categories_data_transformed)
```

```python
new_df = df.drop(['protocol_type','service','flag'], axis=1)
Y = np.array(new_df['label'])
X = pd.DataFrame(np.array(new_df.drop(['label'],axis=1)))
X = pd.concat([X, categories_data_transformed], axis=1)
```

```python
column_names = df.columns
column_names = np.append(np.array(column_names[0]),(np.array(column_names[4:-1])))
column_names = np.append(column_names,np.array(['protocol_type']*3))
column_names = np.append(column_names,np.array(['service']*62))
column_names = np.append(column_names,np.array(['flag']*11))
```

```python
X = np.array(X)
Y[(Y =='back')]=0.0
Y[(Y =='land')]=0.0
Y[(Y =='neptune')]=0.0
Y[(Y =='pod')]=0.0
Y[(Y =='smurf')]=0.0
Y[(Y =='teardrop')]=0.0
Y[(Y !=0.0)]=1.0
Y = Y.astype('float64')
```

```python
#unbalances dataset shapes
print(X.shape)
print(Y.shape)
```

```python
#balancing the dataset
rus = RandomUnderSampler()
X, Y = rus.fit_resample(X, Y)
scaler = preprocessing.StandardScaler().fit(X)
X = scaler.transform(X)
```

```python
#balances dataset shapes
print(X.shape)
print(Y.shape)
```

```python
#Splitting the data into train(80%) and test(20%) randomly 
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state= 42)

#train dataset shapes
print(X_train.shape)
print(Y_train.shape)

#test dataset shapes
print(X_test.shape)
print(Y_test.shape)
```

```python
svc = svm.SVC(kernel="linear",C = 2.8, gamma = .0073,verbose=10)
svc.fit(X_train,Y_train)
train_score = svc.score(X_train,Y_train)
test_score = svc.score(X_test,Y_test)
print("train_score = "+str(train_score))
print("test_score = " +str(test_score))
```

```python
svc = svm.SVC(kernel="rbf",C = 2.8, gamma = .0073,verbose=10)
svc.fit(X_train,Y_train)
train_score = svc.score(X_train,Y_train)
test_score = svc.score(X_test,Y_test)
print("train_score = "+str(train_score))
print("test_score = " +str(test_score))
```

```python
svc = svm.SVC(kernel="poly",C = 2.8, gamma = .0073,verbose=10)
svc.fit(X_train,Y_train)
train_score = svc.score(X_train,Y_train)
test_score = svc.score(X_test,Y_test)
print("train_score = "+str(train_score))
print("test_score = " +str(test_score))
```

```python
svc = svm.SVC(kernel="sigmoid",C = 2.8, gamma = .0073,verbose=10)
svc.fit(X_train,Y_train)
train_score = svc.score(X_train,Y_train)
test_score = svc.score(X_test,Y_test)
print("train_score = "+str(train_score))
print("test_score = " +str(test_score))
```

```python
print("1) I've used 4 different kernals. linear, rbf, poly, sigmoid")
print("2) All the kernals have approximately same train score and test score (> 0.98)")
print("3) But when it comes to number of iterations, poly took more iterations then linear then rbf and least is sigmoid. So the order goes #iteration for poly > #iteration for linear > #iteration for rbf > #iteration for sigmoid")
```

```python
print("The choice a kernel depends on the nature of your data. Here are are the pros and cons of using different kernals in general\n")
print("Linear kernal: ")
print("Pros: 1) Training a linear SVM is often faster and requires less memory, making it suitable for large datasets")
print("      2) Linear kernal generalizes and is less prone to overfitting, which can be a problem with complex kernels.")
print("Cons: 1) If your data is not linearly seperable, using a linear kernel may lead to a model which perform poorly.")
print("\n")
print("RBF kernal: ")
print("Pros: 1) RBF kernal is capable in handeling data which is not linearly seperable also")
print("      2) RBF kernel model is flexible because it's complexity can be controlled by hyperparameters C and gamma.")
print("Cons: 1) Using RBF kernel can be computationally expensive, because of the exponential term")
print("      2) Using RBF kernel may lead to overfitting")
print("\n")
print("Poly kernal: ")
print("Pros: 1) Poly kernal is capable in handeling data which is not linearly seperable also")
print("      2) Poly kernels are generally computationally less expensive than some other non-linear kernels (rbf)")
print("      3) Ploy kernal complexity can be controlled by adjusting the degree of the polynomial")
print("Cons: 1) Using Poly kernal may lead to ovverfitting if we use higher degree polynomial to fit the data")
print("\n")
print("Sigmoid kernal: ")
print("Pros: 1) Sigmoid kernal is capable in handeling data which is not linearly seperable also")
print("      2) Sigmoid kernels are generally computationally less expensive than some other non-linear kernels (rbf)")
print("Cons: 1) Finding the right value for the the hyperparameters alpha and c is not easy")
print("      2) Using Sigmoid kernal may lead to ovverfitting if the hyperparameter are not properly tuned")
```

```python
#diff_srv_rate & dst_host_diff_srv_rate feature
diff_srv_rate_index = np.where(column_names=='diff_srv_rate')[0][0]
dst_host_same_src_port_rate_index = np.where(column_names=='dst_host_diff_srv_rate')[0][0]
X_new = X[:,[diff_srv_rate_index,dst_host_same_src_port_rate_index]]
Y_new = Y
```

```python
#Splitting the data
X_train, X_test, Y_train, Y_test = train_test_split(X_new, Y_new, test_size=0.2, random_state= 42)
```

```python
svc = svm.SVC(kernel="linear")
svc.fit(X_train,Y_train)
train_score = svc.score(X_train,Y_train)
test_score = svc.score(X_test,Y_test)
print("linear kernel")
print("train_score = "+str(train_score))
print("test_score = " +str(test_score))
```

```python
svc = svm.SVC(kernel="rbf")
svc.fit(X_train,Y_train)
train_score = svc.score(X_train,Y_train)
test_score = svc.score(X_test,Y_test)
print("RBF kernel")
print("train_score = "+str(train_score))
print("test_score = " +str(test_score))
```

```python
models = (
    svm.SVC(kernel="linear"),
    svm.SVC(kernel="rbf"),
)
models = (clf.fit(X_train, Y_train) for clf in models)

# title for the plots
titles = (
    "SVC with linear kernel",
    "SVC with RBF kernel"
)

# Set-up 2x2 grid for plotting.
fig, sub = plt.subplots(1, 2)
plt.subplots_adjust(wspace=0.4, hspace=0.4)

X0, X1 = X_train[:, 0], X_train[:, 1]

for clf, title, ax in zip(models, titles, sub.flatten()):
    disp = DecisionBoundaryDisplay.from_estimator(
        clf,
        X_train,
        response_method="predict",
        cmap=plt.cm.coolwarm,
        alpha=0.8,
        ax=ax,
        xlabel='diff_srv_rate',
        ylabel='dst_host_diff_srv_rate',
    )
    ax.scatter(X0, X1, c=Y_train, cmap=plt.cm.coolwarm, s=20, edgecolors="k")
    ax.set_xticks(())
    ax.set_yticks(())
    ax.set_title(title)

plt.show()
```

```python
print("1) I took 'diff_srv_rate' and 'dst_host_diff_srv_rate' as my two features to train the on both linear kernel and RBF kernel")
print("2) As you can see the data is not linearly seperable with these two features")
print("3) Therefore the svm with linear kernal performed very poorly on the data. train and test score less than <0.8")
print("4) But when we use rbf kernal the train and test score are very high (>0.95). The data seems to be linearly seperable in the rbf kernal space")
print("5) Therefore we can use rbf kernal to classify the data using only these two features")
```
