```python
import pandas as pd
import numpy as np
from sklearn import datasets
%matplotlib inline
import matplotlib.pyplot as plt
from matplotlib.colors import ListedColormap
import random
```

```python
def plot_graph(model ,X_1, w, w0):
    plt.scatter(X_1[:40, 0], X_1[:40, 1],
                color='red', marker='o', label='setosa')
    plt.scatter(X_1[40:80, 0], X_1[40:80, 1],
                color='blue', marker='x', label='versicolor')
    
    x1_min, x1_max = X_1[:, 0].min() - 1, X_1[:, 0].max() + 1
    x2_min, x2_max = X_1[:, 1].min() - 1, X_1[:, 1].max() + 1
    xx1, xx2 = np.meshgrid(np.arange(x1_min, x1_max, 0.01), np.arange(x2_min, x2_max, 0.01))
    
    Z = predict(model, np.array([xx1.ravel(), xx2.ravel()]).T, w, w0)
    Z = Z.reshape(xx1.shape)
    
    plt.contourf(xx1, xx2, Z, alpha=0.3, cmap=ListedColormap(['blue', 'red']))
    plt.xlim(xx1.min(), xx1.max())
    plt.ylim(xx2.min(), xx2.max())
    
    plt.show()
    
def predict (model, X, w, w0):
    return np.where((np.dot(X,w) + w0) >= 0.0, 1, -1)

def Perceptron_fit(X, Y, epocs, lr):
    w0 = 0
    w = np.zeros(X.shape[1])
    mistakes = []
    for i in range(epocs):
        mistake =0
        for j in range(X.shape[0]):
            zj = np.where((np.dot(X[j].T,w) + w0) >= 0.0, 1, -1)
            error = Y[j] - zj
            w += lr * error * X[j]
            w0 += lr * error
            mistake += int(error!=0.0)
        mistakes.append(mistake)
            
    return (w, w0, mistakes)

def Adaline_fit(X, Y, epocs, lr):
    w0 = 0
    w = np.zeros(X.shape[1])
    loss_function = []
    for i in range(epocs):
        z = np.dot(X,w) + w0
        error = (Y - z)
        w += (lr * np.dot(X.T, error))
        w0 += (lr * np.sum(error))
        cost = (error**2).sum() / 2.0
        loss_function.append(cost)

    return (w, w0, loss_function) 
```

```python
train_range = []
test_range = []
for i in range(100):
    if i<40:
        train_range.append(i)
    elif 40<=i and i<50:
        test_range.append(i)
    elif 50<=i and i<90:
        train_range.append(i)
    else:
        test_range.append(i)
```

```python
dataset = datasets.load_iris()
X_dataset = np.array(dataset.data)
Y_dataset = np.array(dataset.target)
```

```python
indexes = np.where((Y_dataset == 0) | (Y_dataset == 1))
X_data = X_dataset[indexes]
Y_data = Y_dataset[indexes]
```

```python
# Extract 2 features
X = X_data[:, [0, 2]]
Y = np.where(Y_data==0,-1,1)

# Split the data
X_train = X[train_range,:]
Y_train = Y[train_range]
X_test = X[test_range,:]
Y_test = Y[test_range]
```

```python
# these flowers are seperable using only two features
plt.scatter(X[:50, 0], X[:50, 1],
            color='red', marker='o', label='Setosa')
plt.scatter(X[50:100, 0], X[50:100, 1],
            color='blue', marker='x', label='Versicolor')

plt.xlabel('sepal length [cm]')
plt.ylabel('petal length [cm]')
plt.legend(loc='upper left')

plt.show()
```

```python
## Perceptron
#train
(w, w0, mistakes) = Perceptron_fit(X_train, Y_train, 100, 0.1)

#test
y_pred = predict("per", X_test, w, w0)
print("False predictions on test data (Perceptron) = " + str(np.sum(np.abs(y_pred-Y_test))))
print("Test Accuracy = " + str(100.0*(1 - (np.sum(np.abs(y_pred-Y_test))/len(y_pred))))+"%")

plt.title("Perceptron:")
plot_graph("per", X_train, w, w0)

#plot of number of error predictions in each epocs
plt.title("Error count:")
plt.plot(range(1, len(mistakes) + 1), mistakes, marker='o')
plt.xlabel('Epochs')
plt.ylabel('Number of Errors')
plt.show()
```

```python
## Adaline
#train
(w, w0, loss_function) = Adaline_fit(X_train,Y_train,1000,0.0001)

#test
y_pred = predict("ada", X_test, w, w0)
print("False predictions on test data (Adaline) = " + str(np.sum(np.abs(y_pred - Y_test))))
print("Test Accuracy = " + str(100.0*(1 - (np.sum(np.abs(y_pred-Y_test))/len(y_pred))))+"%")

plt.title("Adaline:")
plot_graph("ada", X_train, w, w0)

#plot of loss function in each epocs
plt.title("Loss Function:")
plt.plot(range(1, len(loss_function) + 1), loss_function, marker='o')
plt.xlabel('Epochs')
plt.ylabel('Loss Function')
plt.show()
```

```python
print("1) If we select 2 features, Both perceptron and Adaline models predicted the test data with 100% accuracy.")
print("2) The Perceptron model gives some fit to the training data. Using two features we are able to get 100% accuracy on test data")
print("3) Adaline model gives the best fit to the training data. Using two features we are able to get 100% accuracy on test data")
print("4) We can see the loss function for both perceptron and Adaline models decrease as the number of epochs increases.")
```
