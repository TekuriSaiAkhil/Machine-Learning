```python
import numpy as np
import pandas as pd
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
from sklearn import tree
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('DataSetForPhishingVSBenignUrl.csv')
df = df.loc[df['URL_Type_obf_Type'].isin(['benign','phishing'])]
df = df.drop_duplicates()
df = df.dropna() 
```

```python
column_names = df.columns[:-1]
X = np.array(df.drop(['URL_Type_obf_Type'], axis=1))
Y = np.array(df['URL_Type_obf_Type'])
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.2, random_state = 42)
```

```python
depth_list = [1, 2, 3, 4, 5, 6]
```

```python
print("Gini criterion:")
for depth in depth_list:
    tree_gini = DecisionTreeClassifier(criterion = "gini", max_depth=depth)
    tree_gini.fit(X_train, Y_train)
    train_score = tree_gini.score(X_train, Y_train)
    test_score = tree_gini.score(X_test, Y_test)
    print("    For depth = "+str(depth)+" (train_score, test_score) = "+"("+str(train_score)+", "+str(test_score)+")")
```

```python
print("Entropy criterion:")
for depth in depth_list:
    tree_entropy = DecisionTreeClassifier(criterion = "entropy", max_depth=depth)
    tree_entropy.fit(X_train, Y_train)
    train_score = tree_entropy.score(X_train, Y_train)
    test_score = tree_entropy.score(X_test, Y_test)
    print("    For depth = "+str(depth)+" (train_score, test_score) = "+"("+str(train_score)+", "+str(test_score)+")")
```

```python
# Discuss your results
print("1) When Gini Impurity measure is used, as the depth of the tree increases the both train accuracy and test accuracy increased")
print("2) When Entropy Impurity measure is used, as the depth of the tree increases the both train accuracy and test accuracy increased")
print("3) This is expected in both imputiry measures because as depth increases, the model complexity increases and able to fit the data more accurately")
print("4) We can also observe in both imputiry that we don't see any overfitting for any depth")
print("5) For each value of depth, both Gini Impurity measure and Entropy Impurity measure resulted in same level of accuracies")
```

```python
from sklearn.tree import export_graphviz
# Export as dot file

tree_entropy = DecisionTreeClassifier(criterion = "entropy", max_depth=2)
tree_entropy.fit(X_train, Y_train)

export_graphviz(tree_entropy, out_file='tree.dot', 
                feature_names = column_names,
                class_names = ['benign','phishing'],
                rounded = True, proportion = False, 
                precision = 2, filled = True)

# Convert to png
from subprocess import call
call(['dot', '-Tpng', 'tree.dot', '-o', 'tree.png', '-Gdpi=600'])

# Display in python
import matplotlib.pyplot as plt
plt.figure(figsize = (14, 18))
plt.imshow(plt.imread('tree.png'))
plt.axis('off');
plt.show();
```
